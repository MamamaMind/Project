```python
#Value counts for categorical features
categorical_columns = ['exercise', 'sugar_intake', 'smoking', 'alcohol', 'married', 'profession', 'health_risk']
for col in categorical_columns:
    if col in data.columns:
        print(f"\nValue counts for {col}:")
        print(data[col].value_counts())
```

```python
#Distribution plots for numeric columns
numeric_cols = ['age', 'weight', 'height', 'sleep', 'bmi']
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Countplots for categorical features
categorical_features = ['exercise', 'sugar_intake', 'smoking', 'alcohol', 'married']

palette = {'high': 'red', 'low': 'green'}

for col in categorical_features:
    plt.figure(figsize=(6,4))
    ax = sns.countplot(x=col, hue='health_risk', data=data, palette=palette)
    
    # Add bar labels
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=1, fontsize=10)
        
    
    plt.title(f'{col.capitalize()} vs Health Risk', fontsize = 14, weight = 'bold')
    plt.xlabel(col.capitalize(), fontsize = 12, weight = 'bold')
    plt.ylabel('Count', fontsize = 12, weight = 'bold')
    plt.legend(title='Health Risk')
    plt.tight_layout()
    plt.show()
```

```python
# Boxplots for numeric vs target
palette = {'high': 'red', 'low': 'green'}

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='health_risk', y=col, data=data, palette=palette)
    
    plt.title(f'{col.capitalize()} by Health Risk', fontsize=14, weight='bold')
    plt.xlabel('Health Risk', fontsize=12, weight='bold')
    plt.ylabel(col.capitalize(), fontsize=12, weight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Mean comparison by health_risk
print("\n=== Average values grouped by health risk ===")
print(data.groupby('health_risk')[numeric_cols].mean())
```

```python
#Correlation heatmap
numeric_cols = ['age', 'weight', 'height', 'sleep', 'bmi']

plt.figure(figsize=(6,5))
corr = data[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```

```python
# Convert categorical columns to numeric 
data_encoded = data.copy()


ordinal_map = {"low": 0, "high": 1}
data_encoded["exercise"] = data_encoded["exercise"].map(ordinal_map)
data_encoded["sugar_intake"] = data_encoded["sugar_intake"].map(ordinal_map)
data_encoded["health_risk"] = data_encoded["health_risk"].map(ordinal_map)

# Convert Yes / No to binary
binary_map = {"yes": 1, "no": 0}
data_encoded["smoking"] = data_encoded["smoking"].map(binary_map)
data_encoded["alcohol"] = data_encoded["alcohol"].map(binary_map)
data_encoded["married"] = data_encoded["married"].map(binary_map)

# Drop columns not suitable for correlation
data_encoded = data_encoded.drop(columns=["profession"])

print(data_encoded.head())

corr = data_encoded.corr()
print(corr)
```

```python
# --- Mean comparison by health risk ---
print("\n Mean Comparison by Health Risk:")
mean_comparison = data_encoded.groupby("health_risk").mean(numeric_only=True)
print(mean_comparison)

# Separate numeric vs binary features
numeric_features = ["age", "weight", "height", "sleep", "bmi"]
binary_features = ["exercise", "sugar_intake", "smoking", "alcohol", "married"]

# --- Plot Numeric Features ---
ax1 = mean_comparison[numeric_features].T.plot(
    kind='bar',
    figsize=(8, 4),
    color=["red", "green"]
)
plt.title("Mean Comparison (Numeric Features) by Health Risk", fontsize = 14, weight = 'bold')
plt.ylabel("Mean Value", fontsize = 12, weight = 'bold')
plt.xlabel("Features", fontsize = 12, weight = 'bold')
plt.legend(["Low Risk (0)", "High Risk (1)"])
plt.xticks(rotation=45)

# Add value labels
for container in ax1.containers:
    ax1.bar_label(container, fmt="%.2f", label_type='edge', fontsize=8)

plt.tight_layout()
plt.show()

# --- Plot Behavioral / Binary Features ---
ax2 = mean_comparison[binary_features].T.plot(
    kind='bar',
    figsize=(8, 4),
    color=["red", "green"]
)
plt.title("Mean Comparison (Behavioral Features) by Health Risk", fontsize = 14, weight = 'bold')
plt.ylabel("Mean Value (0 = No/Low, 1 = Yes/High)", fontsize = 12, weight = 'bold')
plt.xlabel("Features", fontsize = 12, weight = 'bold')
plt.legend(["Low Risk (0)", "High Risk (1)"])
plt.xticks(rotation=45)

# Add value labels
for container in ax2.containers:
    ax2.bar_label(container, fmt="%.2f", label_type='edge', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Lifestyle & Health)")
plt.show()
```

```python
corr_health = corr["health_risk"].sort_values(ascending=False)
print(corr_health)
```

```python
top_corr = corr["health_risk"].drop("health_risk").sort_values(ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x=top_corr.values, y=top_corr.index, palette="coolwarm")
plt.title("Top Correlations with Health Risk")
plt.xlabel("Correlation Value")
plt.ylabel("Feature")
plt.show()
```

## Summary
- Age (0.38), BMI (0.36), and Weight (0.33) have the strongest positive correlation with health risk — meaning older individuals with higher BMI and weight tend to have greater health risk.
- Smoking (0.25) and Alcohol (0.18) also show positive correlations, suggesting lifestyle habits impact health.
- Exercise (-0.19) and Height (-0.18) have negative correlations — meaning more exercise and greater height are slightly associated with lower health risk.
- Other factors like sleep and married status have weaker correlations.
